```python
rhoDM=0.4*gev2g #g/cm^3
cm = plt.cm.get_cmap('RdYlBu')
v0=7.33e-4 ; vesc=2.0e-3  # v0=220 km/s, vesc=500-600 km/s
vstar = np.array([0,0,0])*2e7/c


Norm = lambda v0: np.pi**1.5*v0**3*( scipy.special.erf(vesc/v0)-
                                        (2/np.sqrt(np.pi))*(vesc/v0)*np.exp(-vesc**2/v0**2) ) 

def f_vDM(vx,vy,vz):
    #return np.exp(-v**2/v0**2)*np.heaviside(vesc-v,0.5)/Norm(v0)
    vxrel=vx-vstar[0];
    vyrel=vy-vstar[1];
    vzrel=vz-vstar[2];
    vrel=np.sqrt((vxrel**2 + vyrel**2 + vzrel**2))
    dist= np.exp(-vrel**2/v0**2)*np.heaviside(vesc-vrel,0)
    return dist/Norm(v0)
def f_vDM(v):
    dist= 4*np.pi*v**2*np.exp(-v**2/v0**2)*np.heaviside(vesc-v,0)
    return dist/Norm(v0)
#Norm=nquad(f_vDM, [[  -vesc,vesc],[-vesc,vesc],[  -vesc,vesc] ] ,opts={'epsabs':0,'epsrel':1e-2}
#Norm=nquad(f_vDM, [[  0,vesc]] ,opts={'epsabs':0,'epsrel':1e-2})
#print(Norm)
print(f_vDM(1e-3))
```
